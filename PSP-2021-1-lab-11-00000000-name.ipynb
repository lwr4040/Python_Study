##### 문제 9.
"student_records_2.csv" 파일을 읽고 아래 문제들을 풀어라.

````python
scores = pd.read_csv('student_records_2.csv')
scores
````

##### 9.1
DataFrame scores의 index를 첫번째 열인 '이름'으로 수정하라.

````python
scores.set_index('이름', inplace=True)
scores
````

##### 9.2

````python
len(set(scores['소속']))
````

````python
score_grp_dept.mean()
````

##### 9.5
scores에  각 학생의 점수 합계를 기록한 새로운 column '합계'를 추가하라.

````python
scores['합계'] = scores[['시험1', '시험2', '시험3', '시험4', '시험5']].sum(axis=1)
scores
````

##### 4.6
'합계'의 내림차순으로 scores를 정렬하라.

````python
scores.sort_values(by='합계', ascending=False, inplace=True)
scores
````

##### 9.7
 각 소속별 학생들의 총점의 평균 점수를 막대 그래프로 나타내라.

````python
total_scores = scores[['소속', '합계']]
total_mean_per_dept = total_scores.groupby('소속').mean()
total_mean_per_dept
````

##### 9.8
 학생 수가 가장 많은 소속의 학생들의 정보를 화면에 표시하라.

````python
scores.groupby('소속').size().argmax()
````

````python
max_dept = scores.groupby('소속').size().idxmax()
scores[scores['소속'] == max_dept]
````
