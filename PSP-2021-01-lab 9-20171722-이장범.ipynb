#### 문제 1. 3원색 원 그리기

##### 1.1) 다음 cell에서 img1, img2, img3의 shape을 설명하고, 그 안에 어떤 데이터가 있는지 설명하라.

````python
cols = 201
rows = 201
img1 = np.full((rows,cols,3), 0, np.uint8)
img2 = np.full((rows,cols,3), 0, np.uint8)
img3 = np.full((rows,cols,3), 0, np.uint8)
````

**답**:cols, rows의 img1은 (201, 201, 3)이다.
img2, img3도 다 같다. 
0이 세 개 들어가있는 리스트의 묶음이 들어갔다.

##### 2.2) 다음 cell에서 idx_x, idx_y의 shape을 설명하고, 그 안에 어떤 데이터가 있는지 설명하라.

````python
cr = np.array([65, 65])
cg = np.array([65, 135])
cb = np.array([126, 100])
r = 50

idx_x = np.arange(cols).reshape(-1,1)
idx_y = np.arange(rows).reshape(1,-1)
print(idx_x.shape, idx_y.shape)
````

**답**:idx_x의 shape는 201개의 행을 가지고 있고, 한 개의 열을 가지고 있다.
idx_y의 shape는 201개의 열을 가지고 있고, 한 개의 행을 가지고 있다.
데이터는 0부터 200까지 201개이다.

##### 2.3) 다음 cell에서 img1 안에 어떤 데이터가 있는지 설명하라.

````python
img1[(idx_x-cr[0])**2 + (idx_y-cr[1])**2 < r**2] = np.array([255, 0, 0])
````

````python
plt.imshow(img1)
plt.show()
````

**답**: 3차원 배열이기는 하지만 2차원 배열의 원소가 1차원으로 설명할 수 있다. 각 배열의 원소가 스칼라 값인데 1차원 배열이 2차원 배열로 늘어난 것이다. 원주율이 반지름 안에 있다면 값을 np.array([255,0,0])이 들어가 있다.

##### 2.4) 다음 cell에서 img2와 img3 안에 각각 어떤 데이터가 있는지 설명하라.

````python
img2 = np.full_like(img1, (0, 0, 0))
img3 = np.full_like(img1, (0, 0, 0))
````

````python
img2[(idx_x-cg[0])**2 + (idx_y-cg[1])**2 < r**2] = np.array([0, 255, 0])
````

####  문제 11.

"nature-3082832_640.jpg" 이미지의 흑백 이미지를 $f(x,y)$라고 할 때, 다음 변환을 통해서 새로운 이미지 $g(x,y)$를 만들어 화면에 표시하라.

$$g(x,y) = \begin{cases}
0, & f(x,y) < 52 \\
255/150(f(x,y) - 52), & 52 \le f(x, y) \le 202 \\
0, &f(x, y) > 202
\end{cases}
$$

````python
for i in range(len(img_gray2)):
    for j in range(len(img_gray2[i])):
        if img_gray2[i][j]<52:
            img_gray2[i][j]=0
        elif img_gray2[i][j]>=52 and img_gray2[i][j]<=202:
            img_gray2[i][j]=255/150*(img_gray2[i][j]-52)
        elif img_gray2[i][j]>202:
            img_gray2[i][j]=0
plt.imshow(img_gray2)
plt.show()
````
