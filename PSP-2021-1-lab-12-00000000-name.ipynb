#### 예제 2. Iris 데이터 처리 예

````python
iris = datasets.load_iris(as_frame=True)
````

[Iris Dataset](https://en.wikipedia.org/wiki/Iris_flower_data_set)

````python
dir(iris)
````

````python
df_iris = iris.data
````

````python
iris['target_names']
````

````python
df_iris['sepal length (cm)'].iloc[np.where(iris.target == 0)[0]].hist(bins=30)
````

````python
for class_no in iris.target.unique():
    df_iris['sepal length (cm)'].iloc[np.where(iris.target == class_no)[0]].hist(bins=30)
````

````python
for class_no in iris.target.unique():
    df_iris['sepal width (cm)'].iloc[np.where(iris.target == class_no)[0]].hist(bins=30)
````

````python
for class_no in iris.target.unique():
    df_iris['petal length (cm)'].iloc[np.where(iris.target == class_no)[0]].hist(bins=30)
````

````python
for class_no in iris.target.unique():
    df_iris['petal width (cm)'].iloc[np.where(iris.target == class_no)[0]].hist(bins=30)
````

````python
%matplotlib inline
plt.figure(figsize=(8,6))
for class_no in iris.target.unique():
    plt.scatter(df_iris[iris.target==class_no]['sepal width (cm)'],
                df_iris[iris.target==class_no]['sepal length (cm)'])
#                y='sepal length (cm)')#, c=iris.target)
````

````python
%matplotlib inline
plt.figure(figsize=(8,6))
for class_no in iris.target.unique():
    plt.scatter(df_iris[iris.target==class_no]['petal width (cm)'],
                df_iris[iris.target==class_no]['petal length (cm)'])
#                y='sepal length (cm)')#, c=iris.target)
````

````python
import seaborn as sns
````

````python
sns.histplot(df_iris, x="sepal length (cm)", kde=True)
````

````python
df_iris["class"] = iris.target
````

````python
plt.figure(figsize=(12,10))
plt.subplot(2,2,1)
sns.boxplot(data=df_iris, x="class", y="sepal length (cm)")
plt.xticks([0, 1, 2], iris.target_names)
plt.subplot(2,2,2)
sns.boxplot(data=df_iris, x="class", y="sepal width (cm)")
plt.xticks([0, 1, 2], iris.target_names)
plt.subplot(2,2,3)
sns.boxplot(data=df_iris, x="class", y="petal length (cm)")
plt.xticks([0, 1, 2], iris.target_names)
plt.subplot(2,2,4)
sns.boxplot(data=df_iris, x="class", y="petal width (cm)")
plt.xticks([0, 1, 2], iris.target_names)
plt.show()
````
