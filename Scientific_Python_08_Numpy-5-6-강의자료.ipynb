##### 타원을 이용해서 raccoon의 얼굴 부분만 남기고 검은색으로 변경하기

- 타원의 방정식
$$
\frac{(x-x_c)^2}{a^2} + \frac{(y-y_c)^2}{b^2} = 1
$$

- 타원의 외부
$$
\frac{(x-x_c)^2}{a^2} + \frac{(y-y_c)^2}{b^2} > 1
$$

- 영상에서 왼쪽 위 모서리의 좌표가 $(0, 0)$이고, 세로축의 아래 방향을 x축의 양의 방향, 가로축의 오른쪽 방향을 y축의 양의 방향으로 정의한다.


```python
np.ogrid[0:5, 0:3]
```

```python
X, Y = np.ogrid[0:face.shape[0], 0:face.shape[1]]
X.shape, Y.shape
```

```python
face = misc.face()
center = np.array([350,650])
a = 180
b = 270

idx = (X-center[0])**2/a**2 + (Y-center[1])**2 / b**2 > 0.98
idx &= (X-center[0])**2/a**2 + (Y-center[1])**2 / b**2 < 1.02

face_with_ellipse = face.copy()
face_with_ellipse[idx] = (255, 255, 255)
plt.imshow(face_with_ellipse)
plt.show()
```

```python
center = np.array([350,650])
a = 180
b = 270

idx = (X-center[0])**2/a**2 + (Y-center[1])**2 / b**2 > 1
cropped = face.copy()
cropped[idx] = (0, 0, 0)
plt.imshow(cropped)
plt.show()
```

##### 마름모를 이용해서 raccoon의 얼굴 부분만 남기고 검은색으로 변경하기
- 네 꼭지점의 좌표: $A(350, 380), B(170, 650), C(350, 920), D(530, 650)$
- 두 점 $(x_1, y_1)$과 $(x_2, y_2)$를 잇는 직선의 방정식
$$
y - y_1 = \frac{y_2 - y_1}{x_2 - x_1} (x - x_1)
$$

또는 

$$
y - y_2 = \frac{y_2 - y_1}{x_2 - x_1} (x - x_2)
$$


- 두 점 $(x_1, y_1)$과 $(x_2, y_2)$를 잇는 직선의 윗부분
$$
y - y_1 > \frac{y_2 - y_1}{x_2 - x_1} (x - x_1)
$$

또는 

$$
y - y_2 > \frac{y_2 - y_1}{x_2 - x_1} (x - x_2)
$$


```python
import matplotlib.pyplot as plt
A = np.array([350, 380])
B = np.array([170, 650])
C = np.array([350, 920])
D = np.array([530, 650])
plt.imshow(face)
plt.text(380, 350, "A", bbox=dict(facecolor='white', alpha=0.5))
plt.text(650, 170, "B", bbox=dict(facecolor='white', alpha=0.5))
plt.text(920, 350, "C", bbox=dict(facecolor='white', alpha=0.5))
plt.text(650, 530, "D", bbox=dict(facecolor='white', alpha=0.5))
plt.show()
```

```python
A = np.array([350, 380])
B = np.array([170, 650])
C = np.array([350, 920])
D = np.array([530, 650])
# 직선 AB 윗 부분
slopeAB = (B-A)[1] / (B-A)[0]
idx1 = Y - A[1] < slopeAB * (X - A[0])
img1 = face.copy()
img1[idx1] = (0,0,0)
plt.imshow(img1)
plt.show()
```

```python
# 직선 CD 윗부분
idx2 = Y - C[1] > slopeAB * (X - C[0])
img1[idx2] = (0,0,0)
plt.imshow(img1)
plt.show()
```

```python
# 직선 BC 윗부분
slopeBC = (C-B)[1] / (C-B)[0]
idx3 = Y - C[1] > slopeBC * (X - C[0])
img1[idx3] = (0,0,0)
plt.imshow(img1)
plt.show()
```

```python
# 직선 DA 아랫부분
idx4 = Y - A[1] < slopeBC * (X - A[0])
img1[idx4] = (0,0,0)
plt.imshow(img1)
plt.show()
```

```python
face = misc.face()
A = np.array([350, 380])
B = np.array([170, 650])
C = np.array([350, 920])
D = np.array([530, 650])
# 직선 AB 윗 부분
slopeAB = (B-A)[1] / (B-A)[0]
idx1 = Y - A[1] < slopeAB * (X - A[0])
idx2 = Y - C[1] > slopeAB * (X - C[0])
slopeBC = (C-B)[1] / (C-B)[0]
idx3 = Y - C[1] > slopeBC * (X - C[0])
idx4 = Y - A[1] < slopeBC * (X - A[0])
idx = idx1 & idx2 & idx3 & idx4
img2 = face.copy()
img2[idx] = (0,0,0)
plt.imshow(img2)
plt.show()
```

#### NumPy의 통계 함수: min, max, median, mean, std, var

```python
import numpy as np
x = np.random.normal(3, 0.5, 5)
print(x)
print(np.min(x))
print(np.max(x))
print(np.median(x))
print(np.mean(x))
print(np.std(x))
print(np.var(x))
```
